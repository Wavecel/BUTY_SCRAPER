```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd

df_global=pd.DataFrame(columns=['Product name','Review title','Ratings','Reviews Content','helpfull'])


urls_next_com = set()

def extractor(link):
    global df_global
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:94.0) Gecko/20100101 Firefox/94.0',
        'Accept-Language': 'en-US, en;q=0.5'
    }


    client = requests.session()



    link = link
    sleep(1.5)

    page = client.get(link,headers=headers)


    soup = bs(page.content,'html.parser')


    title = soup.find_all('a',class_='review-title-content')




    review_title = []
    for i in range(0,len(title)):
        review_title.append(title[i].get_text())




    review_title[:] = [titles.lstrip('\n') for titles in review_title]




    review_title[:] = [titles.rstrip('\n') for titles in review_title]



    rating = soup.find_all('i',class_='review-rating')


    rate = []
    for i in range(0,len(review_title)):
        rate.append(rating[i].get_text())



    review = soup.find_all("span",{"data-hook":"review-body"})
  

    review_content = []
    for i in range(0,len(review_title)):
        review_content.append(review[i].get_text())




    review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
   


    review_content[:] = [reviews.rstrip('\n') for reviews in review_content]

    


    df = pd.DataFrame(columns=['Product name','Review title','Ratings','Reviews Content','helpfull'],index=list(range(1,len(review_content) )))






    product_name = soup.find_all("a",{"data-hook":"product-link"})
    product_content =[]
    for i in range(0,len(review_title)):
        product_content .append(product_name[0].get_text())


    



    helpfull = soup.find_all("span",{"data-hook":"review-voting-widget"})


    help_full = []
    for i in range(0,len(helpfull)):
        help_full.append(helpfull[i].get_text())


    help_full[:] = [helpfull1.lstrip('\n') for helpfull1 in help_full ]


    help_full[:] = [helpfull1.rstrip('\n') for helpfull1 in help_full ]


    for i in range(0,len(help_full)):
    #     str1 = "mystring"
    # list1 = list(str1)
    # list1[5] = 'u'
    # str1 = ''.join(list1)
        str1=help_full[i]
        try:
            if str1[0]==' ':
                list1=list(str1)
                list1[0]='0'
                str1=''.join(list1)
                help_full[i]=str1
        except IndexError:
            help_full[i]='0'




    new_item=[]
    for i in range(0,len(help_full)):
        str1=help_full[i]
        str2=str1.split(" ", 1)[0]
        new_item.append(str2)
    f = []
    s = []
    next_rev = soup.find_all('a',href = True)

    for i in range(len(next_rev)): 
        f.append(next_rev[i].get("href"))
    for k in f:
        if "pageNumber" in k:
            s.append(k)

 
    if len(s)>0:
        nxt_url = 'https://amazon.com'+str(s[len(s)-1])
    else:
        nxt_url = 'https://amazon.com'
    
    
    for i in range(0,len(review_content)):
        df['Product name'][i] = product_content[i]
        df['Review title'][i]=review_title[i]
        # print(product_content)
        # print(review_title)
        # print(len(review_title))
        # print(len(rate))
        # print(rate)
        df['Ratings'][i]=rate[i]
        df['Reviews Content'][i]=review_content[i]
        df['helpfull'][i] = new_item[i]
    print(new_item)
    if nxt_url in urls_next_com:
        print("iam done!!!!!!!!!!!!")
    else:
        sleep(1.5)
        urls_next_com.add(nxt_url)
        df_global=pd.concat([df_global,df],axis=0)
        extractor(nxt_url)
        print('iammmmm')
    
    return df

```
